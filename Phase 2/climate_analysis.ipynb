```python
import numpy as np
import pandas as pd
```

```python
daily_data = pd.read_csv('/content/daily_data.csv')
daily_data
```

```python
average_daily_temperature = daily_data.groupby('DATE')['DailyAverageDryBulbTemperature'].mean()
print("Average Daily Temperature:")
print(average_daily_temperature)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(daily_data['DATE'], daily_data['DailySustainedWindSpeed'], label='DailyPeakWindSpeed')
plt.title('Daily Precipitation Trends')
plt.xlabel('Date')
plt.ylabel('DailySustainedWindSpeed (inches)')
plt.legend()
plt.show()
```
