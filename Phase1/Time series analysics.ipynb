# **Time Series Analysics**

```python
import pandas as pd
from prophet import Prophet
from matplotlib import pyplot
from matplotlib.pyplot import figure
from sklearn.metrics import mean_absolute_error
import plotly.express as px
import plotly.graph_objects as go
```

```python
df=pd.read_csv('/content/long_data_.csv')
df
```

```python
df.head()
```

```python
df.shape
```

```python
df['Dates'] = pd.to_datetime(df['Dates'])
```

```python
df = df.groupby(df['Dates'], as_index = False).mean()
df.head()
```

```python
df.shape
```

```python
df = df[['Dates','Usage']]
```

```python
fig = px.line(df, x='Dates', y='Usage')
fig.show()
```

```python
df.columns = ['ds','y']
```

```python
df.head()
```

```python
model=Prophet()
```

```python
model.fit(df)
```

```python
model.component_modes
```

```python
future_dates = model.make_future_dataframe(periods=365,freq='d',include_history=True)
future_dates.shape
```

```python
future_dates.head()
```

```python
prediction=model.predict(future_dates)
```

```python
prediction.head()
```

```python
import plotly.graph_objects as go

trace_open = go.Scatter(
    x=prediction["ds"],
    y=prediction["yhat"],
    mode='lines',
    name="Forecast"
)

trace_high = go.Scatter(
    x=prediction["ds"],
    y=prediction["yhat_upper"],
    mode='lines',
    fill="tonexty",
    line={"color": "#57b8ff"},
    name="Higher uncertainty interval"
)

trace_low = go.Scatter(
    x=prediction["ds"],
    y=prediction["yhat_lower"],
    mode='lines',
    fill="tonexty",
    line={"color": "#ff8c57"},
    name="Lower uncertainty interval"
)

trace_close = go.Scatter(
    x=df["ds"],
    y=df["y"],
    mode='markers',
    name="Data values"
)

# Make a list for all three scatter objects.
data = [trace_open, trace_high, trace_low, trace_close]

# Construct a new Layout object.
layout = go.Layout(
    title="Power Consumption Forecasting",
    xaxis=dict(title="Date"),
    yaxis=dict(title="Power Consumption"),
    legend=dict(x=0, y=1, traceorder="normal", orientation="h"),
    plot_bgcolor="lightgray",  # Set a background color for better visibility
)

fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
fig = go.Figure([go.Scatter(x=df['ds'], y=df['y'],mode='lines',
                    name='Actual')])

fig.add_trace(go.Scatter(x=prediction['ds'], y=prediction['yhat'],
                   mode='lines+markers',
                    name='predicted'))
fig.show()
```

# Optimizing the model for better forecasting

```python
model1=Prophet(daily_seasonality=True).add_seasonality(name='yearly',period=365,fourier_order=70)
```

```python
model1.fit(df)
```

```python
model1.component_modes
```

```python
future_dates1=model1.make_future_dataframe(periods=365)
```

```python
prediction1=model1.predict(future_dates1)
```

```python
# Assuming you have df and prediction1 defined earlier

# Extract true and predicted values
y_true = df['y'].values
y_pred = prediction1['yhat'][:498].values

# Calculate Mean Absolute Error
mae = mean_absolute_error(y_true, y_pred)

# Print the result with additional information
print('Mean Absolute Error (MAE): %.3f' % mae)

# You can also print the actual and predicted values for more context
print('Actual values:', y_true[:5])  # Print the first 5 actual values as an example
print('Predicted values:', y_pred[:5])  # Print the first 5 predicted values as an example
```

```python
# Calculate Mean Absolute Error
y_true = df['y'].values
y_pred = prediction1['yhat'][:498].values
mae = mean_absolute_error(y_true, y_pred)

# Create traces for actual and predicted data
trace_actual = go.Scatter(
    x=df['ds'],
    y=df['y'],
    mode='lines',
    name='Actual'
)

trace_predicted = go.Scatter(
    x=prediction1['ds'],
    y=prediction1['yhat'],
    mode='lines+markers',
    name='Predicted'
)

# Make a list for both traces
data = [trace_actual, trace_predicted]

# Construct a new Layout object
layout = go.Layout(
    title="Time Series Comparison: Actual vs Predicted",
    xaxis=dict(title="Date"),
    yaxis=dict(title="Value"),
    legend=dict(x=0, y=1, traceorder="normal", orientation="h"),
    plot_bgcolor="lightgray",
)

# Create a Figure with data and layout
fig = go.Figure(data=data, layout=layout)

# Add MAE information to the layout
fig.add_annotation(
    text=f'MAE: {mae:.3f}',
    x=df['ds'].iloc[-1],  # Adjust the x-coordinate for annotation placement
    y=y_true.max(),  # Adjust the y-coordinate for annotation placement
    showarrow=True,
    arrowhead=4,
)

# Show the plot
fig.show()
```
